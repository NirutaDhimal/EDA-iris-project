#### 3.2 CDF

```python
class_1 = iris.loc[iris["species"] == 'setosa']
class_2 = iris.loc[iris["species"] == 'versicolor']
class_3 = iris.loc[iris["species"] == 'virginica']
```

```python
#sepal_length corresponding to class_1

counts, bin_edges = np.histogram(class_1['sepal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], cdf, label = 'cdf')
plt.xlabel('sepal length')
plt.title("[CDF PLOT 1.1] Sepal length feature corresponding to class 1")
plt.legend()
plt.show()
```

Conclusion
1. 75 % of flowers with sepal length less than or equal to 5.2 cm are in setosa species.

#### 3.3 Box Plot

```python
#Box plot using feature sepal_length
sns.boxplot(x='species',y='sepal_length', data=iris)
plt.grid()
plt.title("Box plot using feature sepal length")
plt.show()
```

Conclusion
1. Clearly from figure setosa class is not overlapping any other class.Hence, we can conclude that iris flower with sepal length between 4.8 cm to 5.2 cm belongs to setosa species.
2. sepal_length feature can be used to distinguish setosa species from other two species.

```python
#Box plot using feature sepal_width
sns.boxplot(x='species',y='sepal_width', data=iris)
plt.grid()
plt.title("Box plot using feature sepal width")
plt.show()
```

```python
#Box plot using feature petal_length
sns.boxplot(x='species',y='petal_length', data=iris)
plt.grid()
plt.title("Box plot using feature petal  length")
plt.show()
```

Conclusion
1. Clearly from figure we can conclude that petal length feature can be used to distinguish the species of iris flower as no classes are overlapping.
2. Iris flower with petal length 1.3 to 1.6 cm belongs to setosa species.
3. Iris flower with petal length 4 to 4.5 cm belongs to versicolor species.
4 Similarly, Iris flower with petal length 5.1 to 5.9 cm belongs to virginica species.

```python
#Box plot using feature petal_width
sns.boxplot(x='species',y='petal_width', data=iris)
plt.grid()
plt.title("Box plot using feature petal width")
plt.show()
```

Conclusion
1. From figure we can see that no classes are overlapping so petal_width feature can be used to distinguish species of iris flower.
